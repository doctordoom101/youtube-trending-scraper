```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
```

```python
# Setup opsi Chrome
options = Options()
options.add_argument("--headless")  # Hapus ini jika ingin melihat browsernya
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Inisialisasi WebDriver (ganti path jika perlu)
driver = webdriver.Chrome(options=options)
```

# start

```python
# navigasi ke halaman youtube trending
url = 'https://www.youtube.com/feed/trending?bp=6gQJRkVleHBsb3Jl'
driver.get(url)
```

```python
body = driver.find_element(By.CSS_SELECTOR, 'body')
body.send_keys(Keys.END)
time.sleep(3)  # Tunggu agar video baru dimuat
```

```python
video_elements = driver.find_elements(By.ID, 'dismissible')
```

```python
elem = video_elements[0]

# Judul video
title = elem.find_element(By.ID, "video-title").text

# URL video
url_video = elem.find_element(By.ID, "video-title").get_attribute("href")

# Nama channel
channel = elem.find_element(By.ID, "channel-name").text

# Jumlah views
metadata = elem.find_elements(By.ID, "metadata-line")
views = metadata[0].text if metadata else None

# Tampilkan
print("Title:", title)
print("URL:", url_video)
print("Channel:", channel)
print("Views:", views)
```

```python
data = []
titles_seen = []

for elem in video_elements:
    try:
        title = elem.find_element(By.ID, "video-title").text
        if title not in titles_seen:
            titles_seen.append(title)
            url_video = elem.find_element(By.ID, "video-title").get_attribute("href")
            channel = elem.find_element(By.ID, "channel-name").text
            metadata = elem.find_elements(By.ID, "metadata-line")
            views = metadata[0].text if metadata else None
            upload_date = metadata[1].text if len(metadata) > 1 else None

            data.append({
                'title_video': title,
                'url_video': url_video,
                'channel_video': channel,
                'total_views_video': views,
            })
    except Exception as e:
        print("Error:", e)

df = pd.DataFrame(data)
df.head()
```
